```python
import PIL.ImageDraw
from PIL import Image, ImageDraw
import face_recognition
import PIL.Image
import PIL.ImageDraw
```

```python
image= face_recognition.load_image_file("images/faces.jpg")
```

```python
faces_encontradas=face_recognition.face_locations(image)
len(faces_encontradas)
```

```python

from PIL import Image, ImageDraw
image_array=PIL.Image.fromarray(image)
for face in faces_encontradas:
    top, left, bottom, right = face
    if left > right:
        left, right = right, left
    if top > bottom:
        top, bottom= bottom,top
    draw = PIL.ImageDraw.Draw(image_array)
    draw.rectangle(
        [left, top, right, bottom],  # Ordem correta: (x_min, y_min, x_max, y_max)
        outline="red",
        width=3
    )

 
```

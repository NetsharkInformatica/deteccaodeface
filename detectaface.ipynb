```python
import cv2
face_parametro= "cascade/frontalface.xml"
classificador_cascade=cv2.CascadeClassifier(face_parametro)
```

```python
img_face1 = cv2.imread("images/faces.jpg")
img_face1_gray=color.rgb2gray(img_face1)
gray_image_1=cv2.cvtColor(img_face1,cv2.COLOR_BGR2GRAY)

import cv2

faces=classificador_cascade.detectMultiScale(gray_image_1,scaleFactor=1.2,minNeighbors=10,minSize=(30,30))
for (x,y,width,height) in faces:
    cv2.rectangle(img_face1,(x,y),(x+ width, y +height),(0,255,0),6)
    cv2.imwrite("images/faces_resultado.jpg",img_face1)
    Image(url="images/faces_resultado.jpg")

```
